```python
import collections
import os
import tensorflow as tf
import re
import numpy as np
```

```python
with open("allbooks.txt","r",encoding="utf-8") as infile:
    corpus=infile.read().lower()
special_chars = ["[!@#$,.;“”?‘]", "%",'[’]']
def string_cleanup(x, notwanted):
    for item in notwanted:
        x = re.sub(item, '', x)    
    return x

corpus=string_cleanup(corpus,special_chars)
corpus=corpus.split()
print(len(list(set(corpus))))
```
