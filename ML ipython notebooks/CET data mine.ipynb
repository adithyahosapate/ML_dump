```python
import pandas as pd
import urllib
from urllib.request import urlopen
url = 'http://karresults.nic.in/resultCET2017.asp'
from bs4 import BeautifulSoup
import requests
```

```python
import pandas as pd
df = pd.DataFrame(columns=('Name', 'RollNo', 'Engineering','Medical / Dental','ISMH','NATA(Architecture) ','Agriculture BSc.',
                          'Veterinary Science','Agriculture BSc. (Practical)','Veterinary (Practical)','B-PHARMA','D-PHARMA',
                          'LATERAL ENTRY B-PHARMA'))
```

```python
def get_data(rollno):
    response = urlopen("http://karresults.nic.in/indexCET2017.asp")
    page_source = response.read()
    
    soup = BeautifulSoup(page_source, 'html.parser')
    table=soup.find("input")
    payload={"frm_tokens":table["value"],"reg":rollno}
    r = requests.post(url, data=payload)
   
    soup = BeautifulSoup(r.content, 'html.parser')
    marks=soup.find_all("td",{"align":"center"})
    if len(marks)==0:
        return None
    results=soup.find_all("font",{"face":"Verdana"}) 
    name_rollno=soup.find_all("span",{"style":"font-weight: bold"})

    return [*[name_rolln.get_text() for name_rolln in name_rollno],*[results[i].get_text().replace("\xa0","") for i in range(1,22,2)]]
   
    
```

```python
index=0
with open("rollnos.txt","r") as f:
    
    roll_nos = f.readlines()
    
    for i in range(1000):
        a=get_data(roll_nos[i].strip('\n'))
        if a is not None:
            df.loc[index]=a
            index+=1
            if i%100==0:
                print(i)
        else:
            continue
        
       
        
        
```
