```python
import urllib
from urllib.request import urlopen
```

```python
response = urlopen("http://josaa.nic.in/Result/result/OpeningClosingRank.aspx")
page_source = response.read()
```

```python

with open("sc_or_cr_2017.txt","wb" ) as fp:
    fp.write(page_source)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page_source, 'html.parser')
```

```python
print(soup.get_text())
```

```python
table=soup.find("table",{"class":"border_table12"})
print(table)
```

```python
import pandas as pd
df = pd.DataFrame(columns=('Institute', 'Program Name', 'Quota','Opening Rank','Closing Rank'))
#df.loc[i]=[cell for cell in cells]
i=0

rows = table.findAll("tr")
for row in rows:  
    cells=row.findAll("td")
    if len(cells)==5:
        df.loc[i]=[cell.get_text().replace("\n","") for cell in cells]
        i+=1
            
    #cells=row.findAll("td")
#print(cols)
```

```python
df.to_csv("sc_or_cr_2017_round_1.csv",header=True)
```
