```python
import numpy as np
from numpy import random
from random import choice
```

```python
with open('allbooks.txt',encoding='utf-8') as fp:
    array=fp.read()  
    #words=list(array)
    words=array.split(' ')
    print(len(words))
    set_of_words=list(set(words))
    print(set_of_words)
    num_to_word=dict((word,i) for word, i in enumerate(set_of_words))
    word_to_num=dict((i,word) for word,i in enumerate(set_of_words))
    set_of_words=[word_to_num[word] for word in set_of_words]
    print(len(set_of_words))
```

```python
words=[word_to_num[word] for word in words]
print(words[:100])
```

```python
markov_matrix=[]

i=0
for index in range(len(set_of_words)):
    markov_matrix.append({})
for word in words[:-1]:
    if words[i+1] in markov_matrix[word]:
        markov_matrix[word][words[i+1]]+=1
    else:
        markov_matrix[word].update({words[i+1]:1})
    i+=1 
print(len(markov_matrix))
   
print(markov_matrix[1:100])        
```

```python
for index in range(len(markov_matrix)):
    total=0
    for key,value in markov_matrix[index].items():
        total+=value
    for key,value in markov_matrix[index].items():
        markov_matrix[index][key]=float(value/total)
print(markov_matrix[12])          
```
